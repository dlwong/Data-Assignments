```python
import pandas as pd
import matplotlib.pyplot as plt
import datetime
```

```python
csv_file = 'Analytics 1 Master View Audience Overview 20190101-20190131.csv'
df = pd.read_csv(csv_file)
```

```python
df.head()
```

```python
df.info()
```

```python
df['Datetime'] = pd.to_datetime(df['Day Index'])
```

```python
df['Week'] = df['Datetime'].dt.week
```

```python
# Pageviews by week for the segment
plt.figure(figsize=(6, 6))
ax = plt.gca()
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.set_title('Pageviews by Week in January 2019 from RI and NY')
plt.ylabel("Pageviews") 
df.groupby('Week').Pageviews.sum().plot(kind='bar', color=['greenyellow','palegreen','limegreen','seagreen','springgreen'])
```

```python
df['Pageviews/Users'] = df['Pageviews']/df['Users']
```

```python
df.head()
```

```python
df.set_index('Datetime')
```

```python
#Pageviews/user by day for the segment
plt.figure(figsize=(6, 6))
plt.xlabel("Day") 
plt.ylabel("Pageviews/user") 
ax = plt.gca()
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.set_title('Pageviews Per User by Day in January 2019 from RI and NY')
df['Pageviews/Users'].plot(color='mediumslateblue')
```
