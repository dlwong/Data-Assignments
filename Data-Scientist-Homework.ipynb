```python
import statsmodels.api as sm

dummy_genders = pd.get_dummies(df['gender'], prefix = 'gender')
dummy_metro = pd.get_dummies(df['metropolitan_area'], prefix = 'metro_area') 
dummy_device = pd.get_dummies(df['device_type'], prefix = 'device')
```

```python
cols_to_keep = ['active', 'age']
activity_data = df[cols_to_keep].join(dummy_genders.ix[:, 'gender_M':]) #only taking male column
activity_data = activity_data.join(dummy_metro.ix[:, 'metro_area_Birmingham':]) #ignored austin
activity_data = activity_data.join(dummy_device.ix[:, 'device_Mobile':]) #ignored desktop
activity_data = sm.add_constant(activity_data, prepend=False)
explanatory_cols = activity_data.columns[1:] #independent variables
full_logit_model = sm.GLM(activity_data['active'],
                    activity_data[explanatory_cols], 
                    family=sm.families.Binomial()) #it serves as a garbage bin for any bias that is not accounted 
                                #for by the terms in the model. 
result = full_logit_model.fit()
#Apply this model to the same data that the model was trained on and assess the prediction accuracy.
```

```python
print(result.summary())
```

```python
probs=result.predict(activity_data[explanatory_cols])
```

```python
import numpy as np

pred = np.where(probs > 0.5, 1, 0)
```

```python
from sklearn.metrics import accuracy_score #accuracy = matches/samples

accuracy_score(activity_data['active'], pred)
```

```python
training_data = activity_data[1:4000]
test_data = activity_data[4001:].copy()
training_logit_model = sm.GLM(training_data['active'],
                        training_data[explanatory_cols],
                        family=sm.families.Binomial()) 
training_result = training_logit_model.fit()
#Assess the training data model’s accuracy on the test data. Why does the accuracy change so much?
```

```python
print(training_result.summary())
# change in the log odds of the outcome for a one unit increase in the predictor variable.
#For every one unit change in male, the log odds of admission decreases by -0.58.
#for living in birmingham instead of other places changes the log odds of active by -0.1187.
```

```python
test_probs=training_result.predict(test_data[explanatory_cols])
training_probs=training_result.predict(training_data[explanatory_cols])
```

```python
test_pred = np.where(test_probs > 0.5, 1, 0)
training_pred = np.where(training_probs > 0.5, 1, 0)
```

```python
accuracy_score(test_data['active'], test_pred)
```

```python
accuracy_score(training_data['active'], training_pred)
```
