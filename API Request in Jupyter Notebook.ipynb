```python
import requests
import yaml
import pandas as pd
```

```python
credentials = yaml.load(open('github.yaml'))
api_key = credentials['GITHUB_KEY']
```

```python
response = requests.get(
    'https://api.github.com/users/dlwong/repos',
    headers={'Authorization': 'token '+api_key}
)
```

```python
df = pd.DataFrame(response.json())
```
